## HATEOAS

- I am covering so that you are familiar with the concept. It may come up in interviews, ...


- "Hypermedia as the Engine of Application State (HATEOAS) is a component of the REST application architecture that distinguishes it from other network application architectures.
<br><br>
With HATEOAS, a client interacts with a network application whose application servers provide information dynamically through hypermedia. A REST client needs little to no prior knowledge about how to interact with an application or server beyond a generic understanding of hypermedia.
<br><br>
By contrast, clients and servers in CORBA interact through a fixed interface shared through documentation or an interface description language (IDL).
<br><br>
The way that the HATEOAS constraint decouples client and server enables the server functionality to evolve independently." (https://en.wikipedia.org/wiki/HATEOAS)


- There are three core concepts:
    - "What did I get?" $\rightarrow$ Content-Type
    - "Where can I go?" $\rightarrow$ Links
    - "What operations do I support?" $\rightarrow$ Methods in links.

# OAuth2

## Concept

- "OAuth is an open standard for access delegation, commonly used as a way for Internet users to grant websites or applications access to their information on other websites but without giving them the passwords.[1] This mechanism is used by companies such as Amazon, Google, Facebook, Microsoft and Twitter to permit the users to share information about their accounts with third party applications or websites." (https://en.wikipedia.org/wiki/OAuth)


- "The OAuth 2.0 authorization framework enables a third-party
   application to obtain limited access to an HTTP service, either on
   behalf of a resource owner by orchestrating an approval interaction
   between the resource owner and the HTTP service, or by allowing the
   third-party application to obtain access on its own behalf." (https://tools.ietf.org/html/rfc6749)
   
   
- "OAuth 2.0 is a protocol that allows a user to grant limited access to their resources on one site, to another site, without having to expose their credentials." (https://auth0.com/docs/protocols/oauth2)


<hr style="height:3px">

| <img src="../../images/oauth2.png" width="900px"> |
| :---: |
| __OAuth2__ |

<hr style="height:3px">


- OAuth2 supports _authentication_ and _authorization._


- Authentication:
    - The "user" typically needs to "logon" to the resource server before granting access.
    - Successful login generates a token passed to the client application.
    - If the token "works," the authorization server has also acted as an authentication server.
    
    
- Authorization:
    - Typically, authorization involved looking up what a user can do in an authorization database.
    - File protections in Unix are an example.
    - In OAuth2, the issued token contains the permissions. This enables the resource server to evaluate the request without performing lookups.
    - You implemented a hard coded authorization "database" for the middleware/security.
    - The authorization rules are not passed on the request.
    
    
- "Authorization: This is the most common scenario for using JWT. Once the user is logged in, each subsequent request will include the JWT, allowing the user to access routes, services, and resources that are permitted with that token. Single Sign On is a feature that widely uses JWT nowadays, because of its small overhead and its ability to be easily used across different domains." (https://jwt.io/introduction/)

    
- A simplest view of a JWT token containing authorization (access information) might look like:


```
{
    "header": {
          "alg": "HS256",
          "typ": "JWT"
    },
    "claims": [
        { 
            "sub": "/api/customers/12345",
            "operations": ["GET", "PUT"]
        },
        { 
            "sub": "/api/customers/1234/profiles",
            "operations": ["POST"]
        },
        { 
            "sub": "/api/profiles?customerid=1234",
            "operations": ["GET", "PUT", "DELETE"]
        }
    ],
    "signature": {
        HMACSHA256(
             base64UrlEncode(header) + "." +
              base64UrlEncode(payload),
              secret)
    }
}
```

- The secret may is not really __in__ the signature. The snippet above is an API call that produces the signature.


- And remember, the whole thing is encrypted.

## Concept

<hr style="height:2px">

- You can clearly implement an orchestration/choreography using programming languages like Python, C++, etc.


- There are also special tools, languages and runtime engines/interpreters that specialize in simplifying the implementation of service choreography/orchestration.


- "Business Process Model and Notation (BPMN) is a graphical representation for specifying business processes in a business process model.
<br><br>
Business Process Model and Notation (BPMN) is a standard for business process modeling that provides a graphical notation for specifying business processes in a Business Process Diagram (BPD), based on a flowcharting technique very similar to activity diagrams from Unified Modeling Language (UML). The objective of BPMN is to support business process management, for both technical users and business users, by providing a notation that is intuitive to business users, yet able to represent complex process semantics.(https://en.wikipedia.org/wiki/Business_Process_Model_and_Notation)"


<hr style="height:2px">

| <img src="../../images/simple_bpmn_diagram.png" width="1000px"> |
| :---: |
| __BPMN__ |

<hr style="height:2px">

- BPMN can be used for collaborative design, similar to the way teams use [UML](https://en.wikipedia.org/wiki/Unified_Modeling_Language) or [entity-relationship modeling](https://en.wikipedia.org/wiki/Entity%E2%80%93relationship_model) tools/languages.


- In the same way that teams use ER or UML tools to _generate_ executable artifacts, teams can use BPMN and related tools to produce executable workflows/orchestration/choreography.


- There are integrated product suites with functionality for:
    - Modeling, definition tools.
    - Engine for executing defined processes.
    - Analysis and reporting on processes and business performance indicators.

- Why would people use service orchestration/choreography instead of just "writing code?" There are several reasons:
    - The orchestration processes run for a long time (hours, days, ...), and task invocations are asynchronous.
        - This type of code is difficult to write.
        - Engines suspend and resume the process when task responses arrive.
    - The approach is more intuitive to some people for some types of applications.
    - The integrated approach to visual orchestration design, execution, tracing and reporting often works well for collaboration between "programmers" and "line of business managers." The latter do not want to understand Python code, look at RDB data, etc.
    
    
- But, coding is still a very common approach.
